# 3) (a) (5 marks) Train a decision tree using both the Gini index and the Entropy by changing the max-depth [4, 8, 10, 15, 20]. Dont change any of the other default values of the classifier. In the following model, use the criteria which give better accuracy on the test set with the chosen depth
importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
reader1=pd.read_csv('BitcoinHeistData.csv')
df1=pd.DataFrame(reader1)
length=len(df1)
df1.describe()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

Handling the null values in the dataset

```python
df1=df1.dropna()
df1.isnull().sum().sum()
```

We have dropped any row that might have had missing values. Now there are no null values in the dataset.

```python
df1.shape
```

The dataset now has 2916697 rows and 10 columns

```python
df1=df1.sample(frac=1,random_state=0).reset_index(drop=True)
df1.head()
```

```python
le=LabelEncoder()
df1['address']=le.fit_transform(df1['address'])
df1['label']=le.fit_transform(df1['label'])
df1.head()
```

```python
df1.info()
```

```python
df2=df1.drop(['label'],axis=1)
df2.head()
```

```python
temp=df1['label']
temp.head()
```

```python
X_train,X_val,X_test=df2[:int(0.7*length)],df2[int(0.7*length):int(0.85*length)],df2[int(0.85*length):]
y_train,y_val,y_test=temp[:int(0.7*length)],temp[int(0.7*length):int(0.85*length)],temp[int(0.85*length):]
```

Training Decision Tree models

```python
max_depth_array=[4,8,10,15,20]
for i in max_depth_array:
    clf1=DecisionTreeClassifier(criterion="gini",max_depth=i,random_state=0)
    clf1.fit(X_train,y_train)
    y_pred=clf1.predict(X_test)
    print("(Testing set)Accuracy for max_depth =",i,"and criterion gini is",accuracy_score(y_test,y_pred))
    y_pred=clf1.predict(X_val) 
    print("(Validation set)Accuracy for max_depth =",i,"and criterion gini is",accuracy_score(y_val,y_pred))
for i in max_depth_array:
    clf2=DecisionTreeClassifier(criterion="entropy",max_depth=i,random_state=0)
    clf2.fit(X_train,y_train)
    y_pred=clf2.predict(X_test)
    print("(Testing set)Accuracy for max_depth =",i,"and criterion entropy is",accuracy_score(y_test,y_pred))
    y_pred=clf2.predict(X_val) 
    print("(Validation set)Accuracy for max_depth =",i,"and criterion entropy is",accuracy_score(y_val,y_pred))
```

We can see that we get the best accuracy score with the criterion entropy with max depth as 15 and the accuracy score is 0.9888458417618085 for the testing set, and for the validation set, we get the best accuracy score with the criterion entropy with max depth as 15 and the accuracy score is 0.9888024136867007

# 3) (b) (5 marks) Ensembling is a method to combine multiple not-so-good models to get a better performing model. Create 100 different decision stumps (max depth 3). For each stump, train it on randomly selected 50% of the training data, i.e., select data for each stump separately. Now, predict the test samples labels by taking a majority vote of the output of the stumps. How is the performance affected as compared to parts (a)
importing libraries

```python
from scipy import stats as st
```

```python
correct=0
X_train_copy=X_train.copy()
y_train_copy=y_train.copy()
train_copy=pd.concat([X_train_copy,y_train_copy],axis=1)
output=[]
output_val=[] 
for i in range(100):
    train_copy=train_copy.sample(frac=1,random_state=i).reset_index(drop=True)
    X_train_copy=train_copy.drop(['label'],axis=1)[:int(0.5*length)]
    y_train_copy=(train_copy['label'])[:int(0.5*length)]
    clf=DecisionTreeClassifier(criterion="entropy",splitter="random",max_depth=3,random_state=i)
    clf.fit(X_train_copy,y_train_copy)
    output.append(clf.predict(X_test))
    output_val.append(clf.predict(X_val)) 
output2=[]
output_val2=[] 
for i in output:
  output2.append(np.array(i))
output2=np.array(output2)
output2=output2.T
print(output2.shape,X_test.shape)
print(output2[0].shape)
print(output2[0])
for i in output_val:
  output_val2.append(np.array(i))
output_val2=np.array(output_val2)
output_val2=output_val2.T
print(output_val2.shape,X_val.shape)
for i in range(0,output2.shape[0]):
  if np.bincount(output2[i]).argmax()==y_test.iloc[i]:
    correct+=1
print("(Testing set)Accuracy is",correct/len(X_test))
correct=0 
for i in range(0,output_val2.shape[0]):
  if np.bincount(output_val2[i]).argmax()==y_val.iloc[i]:
    correct+=1
print("(Validation set)Accuracy is",correct/len(X_val))
```

# 3) (c) (5 marks) Another popular boosting technique is Adaboost. Use the sklearn Adaboost algorithm on the above dataset and report the testing accuracy. Use the Decision tree as the base estimator and with a number of estimators as [4, 8, 10, 15, 20]. Compare RF and Adaboost results.
importing libraries

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
estimators=[4,8,10,15,20]
for i in estimators: 
    clf4=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(criterion="entropy",max_depth=15,random_state=0),n_estimators=i,random_state=0)
    clf4.fit(X_train,y_train)
    y_pred=clf4.predict(X_test)
    print("(Testing set)Accuracy for n_estimators =",i,"is",accuracy_score(y_test,y_pred))
    y_pred=clf4.predict(X_val) 
    print("(Validation set)Accuracy for n_estimators =",i,"is",accuracy_score(y_val,y_pred))
for i in estimators:
    clf5=RandomForestClassifier(criterion="entropy",max_depth=15,n_estimators=i,random_state=0)
    clf5.fit(X_train,y_train)
    y_pred=clf5.predict(X_test)
    print("(Testing set)Accuracy for n_estimators =",i,"is",accuracy_score(y_test,y_pred))
    y_pred=clf5.predict(X_val) 
    print("(Validation set)Accuracy for n_estimators =",i,"is",accuracy_score(y_val,y_pred))
```
